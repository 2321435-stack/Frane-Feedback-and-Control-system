```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl
```

```python
# Define transfer function G(s) = (s+1) / (2s^2 + s)
num = [1, 1]          
den = [2, 1, 0]      
```

```python
G = ctrl.TransferFunction(num, den)

print("Transfer Function G(s):")
print(G)
```

```python
# Step Response
t, y = ctrl.step_response(G)

plt.figure()
plt.plot(t, y)
plt.title("Step Response")
plt.xlabel("Time (s)")
plt.ylabel("Output")
plt.grid()
plt.show()
```

```python
# Pole-Zero Map
plt.figure()
ctrl.pzmap(G, plot=True, grid=True)
plt.title("Pole-Zero Map")
plt.show()
```
