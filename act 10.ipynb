```python
from openai import OpenAI
from langchain_community.chat_message_histories import ChatMessageHistory

# Setup client - ensure LM Studio is running on 1234 with a CHAT model loaded
client = OpenAI(
    base_url="http://127.0.0.1:1234/v1", 
    api_key="lm-studio"
)

history = ChatMessageHistory()

print("Chatbot initialized for Laurence Frane.")
print("Type 'exit' anytime to stop.\n")

while True:
    query = input("User Input: ") # Changed prompt to match your example

    if query.lower() == "exit":
        print("\nExiting...")
        print("Program Finished.")
        break

    if not query.strip():
        continue

    # Header line before processing
    print("-" * 99)

    history.add_user_message(query)

    messages_with_history = []
    for m in history.messages:
        role = "user" if m.type == "human" else "assistant"
        messages_with_history.append({
            "role": role,
            "content": m.content
        })

    try:
        stream = client.chat.completions.create(
            model="model-identifier", 
            messages=messages_with_history,
            stream=True
        )

        full_ai_response = ""Can you give me 2 commonly used microcontrollers, with one short description each?

        # Stream response
        for chunk in stream:
            if chunk.choices[0].delta.content:
                content = chunk.choices[0].delta.content
                print(content, end="", flush=True)
                full_ai_response += content

        history.add_ai_message(full_ai_response)

        # Footer separator to match your example
        print("\n \n" + "=" * 100 + "\n ")

    except Exception as e:
        print(f"\n❌ Error: {e}")
        print("=" * 100 + "\n")
```
