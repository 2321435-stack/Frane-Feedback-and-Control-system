```python
import control as ct
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1: Define the system coefficients
num = [1, 1]
den = [1, 4, 7, 6, 3]
sys = ct.TransferFunction(num, den)

# Step 2: Display the transfer function
print("Transfer Function:")
print(sys)
```

```python
# Step 3: Calculate poles and zeros (Fixed function names)
p = ct.poles(sys)
z = ct.zeros(sys)
print(f"Poles: {p}")
print(f"Zeros: {z}")
```

```python
# Step 4: Plot the step response
t, y = ct.step_response(sys)
plt.figure(figsize=(10, 5))
plt.plot(t, y, color='teal', linewidth=2)
plt.title('Step Response', fontweight='bold')
plt.xlabel('Time (s)')
plt.ylabel('Output')
plt.grid(True, linestyle='--')
plt.show()
```

```python
# Step 5: Plot the pole-zero map
plt.figure(figsize=(6, 6))
ct.pzmap(sys, grid=True)
plt.title('Pole-Zero Map', fontweight='bold')
plt.show()
```
