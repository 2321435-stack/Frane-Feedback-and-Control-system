```python
import control as ct
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1: Set the system coefficients
num = [1]
den = [1, 2, 3, 2, 0]

# Step 2: Build the transfer function
sys = ct.TransferFunction(num, den)
print("Transfer Function:")
print(sys)
```

```python
# Step 3: Find the poles and zeros
p = ct.poles(sys)
z = ct.zeros(sys)
print(f"Poles: {p}")
print(f"Zeros: {z}")
```

```python
# Step 4: Create the step response plot
t, y = ct.step_response(sys)
plt.figure(figsize=(10, 5))
plt.plot(t, y, color='#0d9488', linewidth=2)
plt.title('Step Response', fontweight='bold')
plt.xlabel('Time (s)')
plt.ylabel('Output')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Step 5: Create the pole-zero map

plt.figure(figsize=(8, 6))
ct.pzmap(sys, grid=True)
plt.title('Pole-Zero Map', fontweight='bold')
plt.show()
```
