```python
import control as ct
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1: Define system arrays
# These coefficients match G(s) = 3 / (8s^4 + 12s^3 + 26s^2 + 18s)
num_coeffs = np.array([3])
den_coeffs = np.array([8, 12, 26, 18, 0])

# Step 2: Initialize the Transfer Function object
LTI_system = ct.tf(num_coeffs, den_coeffs)
print("Transfer Function G(s):")
print(LTI_system)
```

```python
# Step 3: Compute critical system points
sys_poles = ct.poles(LTI_system)
sys_zeros = ct.zeros(LTI_system)
print(f"Poles: {sys_poles}")
print(f"Zeros: {sys_zeros}")
```

```python
# Step 4: Step response simulation and plotting
# Generating the time and output data points
time, amplitude = ct.step_response(LTI_system)

plt.figure(figsize=(10, 6))
plt.plot(time, amplitude, color='#0D9488', lw=2.5)
plt.title('System Transient Response', fontweight='bold', loc='left')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(visible=True, linestyle=':', alpha=0.8)
plt.tight_layout()
plt.show()
```

```python
# Step 5: Pole-Zero mapping on the s-plane

plt.figure(figsize=(10, 6))
ct.pzmap(LTI_system, grid=True, title='S-Plane Pole-Zero Distribution')
plt.show()
```
