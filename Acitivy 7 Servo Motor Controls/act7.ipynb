```python
import serial
import time
import sys
```

```python
# Serial configuration
PORT = "COM12"
BAUD = 9600

def connect_arduino():
    try:
        arduino = serial.Serial(PORT, BAUD, timeout=1)
        time.sleep(2)  # wait for Arduino to reset
        print(f"Connected to Arduino on {PORT}")
        return arduino
    except Exception as err:
        print("Error: Could not open serial port:", err)
        sys.exit()
```

```python
def main():
    ser = connect_arduino()

    print("Arduino Servo Controller")
    print("Type 'q' to quit")

    while True:
        user_input = input("Enter degrees (0-180): ").strip()

        if user_input.lower() == "q":
            print("Program ended.")
            ser.close()
            break

        # Send value to Arduino
        ser.write((user_input + "\n").encode())

        time.sleep(0.1)

        # Read feedback from Arduino
        while ser.in_waiting:
            message = ser.readline().decode().strip()
            print(message)
```

```python
if __name__ == "__main__":
    main()
```
